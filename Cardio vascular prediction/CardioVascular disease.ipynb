```python
#featre selection

```

```python
from sklearn.feature_selection import SelectKBest
```

```python
from sklearn.feature_selection import f_classif
```

```python
Fit_feature=SelectKBest(score_func=f_classif)
```

```python
Fit_feature.fit(x,y)
```

```python
score_data=pd.DataFrame(Fit_feature.scores_,columns=['score'])
```

```python
col_name=pd.DataFrame(x.columns)
col_name
```

```python
score=pd.concat([col_name,score_data],axis=1)
```

```python
score
```

```python
score.nlargest(5,score_data)
```

# feature importance

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
model=ExtraTreesClassifier()
```

```python
model.fit(x,y)
```

```python
model.feature_importances_
```

```python
top_feature=pd.Series(model.feature_importances_,index=x.columns)
```

```python
top_feature
```

```python
top_feature.nlargest(10).plot(kind='barh')
```
